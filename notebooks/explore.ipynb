```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
```

```python
file = '../data/ktwo201885041-c14_llc.fits'
```

# How to read a FITS file
(because I can never remember, and always have to refer to the [doccumentation](http://docs.astropy.org/en/stable/io/fits/))

```python
hdu = fits.open(file)
```

```python
hdu.info()
```

```python
lc = hdu[1].data

hdu.close()
```

```python
lc.columns
```

```python
plt.figure(figsize=(14,6))

plt.plot(lc['TIME'], (lc['PDCSAP_FLUX'] / np.nanmedian(lc['PDCSAP_FLUX'])) - 1)

plt.xlabel('BJD - 2454833 (days)')
plt.ylabel('Flux (e-/s)')
plt.title('K2 data for Wolf 359')

plt.ylim(-0.01, 0.06)
plt.xlim(3090,3110)
```
